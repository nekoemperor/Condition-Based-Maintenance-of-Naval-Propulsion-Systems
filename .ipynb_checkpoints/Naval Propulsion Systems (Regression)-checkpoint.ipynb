## Import Train and Test dataset

```python
column_list_file = 'naval_train_column_list.txt'
train_file = 'naval_train.csv'
test_file = 'naval_test.csv'
```

```python
columns = ''
with open(column_list_file,'r') as f:
    columns = f.read().split(',')
```

```python
# Specify the column names as the file does not have column header
df_train = pd.read_csv(train_file,names=columns)
df_test = pd.read_csv(test_file,names=columns)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
X_train = df_train.iloc[:,0:25] # Features: 0 column onwards till 25th column
y1_train = df_train.iloc[:,25].ravel() # Target: 25th column
y2_train = df_train.iloc[:,26].ravel() # Target: 26th column
y3_train = df_train.iloc[:,27].ravel() # Target: 27th column
y4_train = df_train.iloc[:,28].ravel() # Target: 28th column
y5_train = df_train.iloc[:,29].ravel() 

X_test = df_test.iloc[:,0:25]
y1_test = df_test.iloc[:,25].ravel()
y2_test = df_test.iloc[:,26].ravel()
y3_test = df_test.iloc[:,27].ravel()
y4_test = df_test.iloc[:,28].ravel()
y5_test = df_test.iloc[:,29].ravel()
```

```python
X_train
```

```python
y1_train
```

```python
y2_train
```

## Select models

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import Ridge,ElasticNet
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

#MLP
from sklearn.neural_network import MLPRegressor

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
from keras.constraints import maxnorm


pd.plotting.register_matplotlib_converters()

models = []
models.append(('LinearRegression', LinearRegression()))
# # models.append(('PNR', PolynomialFeatures(degree = 4))) # pnr has no score model
models.append(('pipelineSVR', Pipeline([('scalar1',StandardScaler()),
                     ('pca1',PCA(n_components=2)),
                     ('lr_classifier',SVR(kernel='linear'))])))
# models.append(('Ridge', Ridge(alpha=2,max_iter=1000,random_state=0)))
models.append(('RandomForestRegression', RandomForestRegressor(n_estimators = 10, random_state = 0)))
models.append(('GradientBoostingRegressor', GradientBoostingRegressor(random_state=0)))
models.append(('XGBoostRegression', XGBRegressor(verbosity = 0, random_state=0)))
models.append(('DecisionTreeRegreesion', DecisionTreeRegressor(random_state = 0)))
```

## Select best of the best model for y1

```python
for name, model in models:
    if name=='SVR':
        sc = StandardScaler()
        X_trainSVR = sc.fit_transform(X_train)
        y1_trainSVR = sc.fit_transform(y1_train)
        model.fit(X_trainSVR.reshape(-1, 1), y1_trainSVR)
        scores = model.score(X_trainSVR, y1_trainSVR)
        print(name,"Score: {:.2f} %".format(scores*100))
        
    else:
        model.fit(X_train, y1_train)
        scores = model.score(X_train, y1_train)
        print(name,"Score: {:.2f} %".format(scores*100))
#           ,"Standard Deviation: {:.2f} %".format(scores.std()*100))
```

```python
## Highest score for training = DecisionTreeRegression (100% overfitting??)
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y1_train)

# ## 2nd highest score  = RandomForestRegression (99,99% overfitting??)
# regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
# regressor.fit(X_train, y1_train)

# ## 3rd highest score = XGBoostRegression
# regressor = XGBRegressor(verbosity = 0, random_state=0)
# regressor.fit(X_train, y1_train)
```

```python
results = X_test.copy()
results
```

```python
y1_pred = regressor.predict(X_test)

results["predicted"] = y1_pred 
results["actual"] = y1_test

results = results[['predicted', 'actual']]
results['predicted'] = results['predicted'].round(2)
results
```

```python
y1_test
```

```python
y1_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

r2score = r2_score(y1_test, y1_pred)
rms = mean_squared_error(y1_test, y1_pred, squared=False)

print("R2-Score: {:.5f} %".format(r2score*100))
print("RMSE: {:.5f} %".format(rms*100))
```
